# Hệ Thống Gợi Ý Nguyện Vọng Xét Tuyển Đại Học

Notebook demo hệ thống gợi ý ngành học cho thí sinh 2024.

```python
import sys
import os

# Add src to path
sys.path.insert(0, os.path.dirname(os.path.dirname(os.path.abspath('__file__'))))

from src.recsys import load_data, recommend
```

```python
# Load data
data = load_data()
print(f"Loaded {len(data['predictions_2024']):,} predictions")
```

## Gợi ý cơ bản (chưa chọn ngành)

```python
recs = recommend(sbd="1000001", data=data)
display(recs)
```

## Gợi ý với prior selections

```python
# Thí sinh đã chọn các ngành (theo thứ tự NV)
selected = [
    ("QSC", "7340101"),  # NV1
    ("MHN", "7480201"),  # NV2
]

recs = recommend(sbd="1000001", data=data, selected_majors=selected)
display(recs)
```

## Interactive

```python
# ========================================
# NHẬP THÔNG TIN
# ========================================

SBD = "1000001"

# Sắp xếp theo thứ tự NV (NV1 đầu tiên)
SELECTED = [
    # ("mã_trường", "mã_ngành"),
]

SCORE_TOLERANCE = 1.0
TOP_K = 10

# ========================================

recs = recommend(
    sbd=SBD,
    data=data,
    selected_majors=SELECTED if SELECTED else None,
    score_tolerance=SCORE_TOLERANCE,
    top_k=TOP_K
)

if len(recs) > 0:
    display(recs)
```
